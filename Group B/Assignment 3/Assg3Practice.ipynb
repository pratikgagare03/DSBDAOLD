```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Airquality.csv")
df.head(12)
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis=1)
df.head(6)
```

```python
df['Ozone']=df['Ozone'].fillna(df['Ozone'].mean())
df['Solar.R']=df['Solar.R'].fillna(df['Solar.R'].mean())
```

```python
df['Humidity']=df['Humidity'].fillna(df['Humidity'].mode()[0])
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Humidity']=label.fit_transform(df['Humidity'])
df['Humidity'].unique()
```

```python
df.dtypes
```

```python
plt.scatter(df["Ozone"],df["Temp"])
plt.legend(["Ozone"])
plt.xlabel("Ozone")
plt.ylabel("Temp")
plt.title("Ozone Scatter Plot")
plt.show()
```

```python
import seaborn as sns
sns.barplot(x=df["Month"], y=df["Temp"])
```

```python
corr = df.corr()
sns.heatmap(corr, annot = True)
```

```python
sns.pairplot(df)
```

```python
h = df.iloc[1:15, 0]
v = df.iloc[1:15, 3]
plt.plot(h, label="Ozone", marker="o")
plt.plot(v, label="Temp", marker="o")
plt.title("Line Graph for Ozone and Temp")
plt.legend()
plt.show()
```

```python
labels= ['Ozone','Solar.R','Wind','Temperature']
sizes=[df['Ozone'].mean(),df["Solar.R"].mean(),df['Wind'].mean(),df["Temp"].mean()]
colors=['red','pink','yellow','silver']
textprops = {"fontsize":15}
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title("Airquality Factors", fontsize=20, style="italic", pad=35)
```

```python
plt.pie([df["Ozone"].mean(),df["Solar.R"].mean(),df["Wind"].mean(),df["Temp"].mean()], autopct="%1.0f%%", labels=["Ozone", "Solar.R", "Wind", "Temp"])
plt.plot()
```

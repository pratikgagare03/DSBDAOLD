## Importing libraries and reading dataset

```python
import pandas as pd
import numpy as np
```

## f) Creating subset using iloc 

```python
subset1=data.iloc[:,:8]
subset1
```

## g) Shape of subset

```python
subset1.shape
```

## h) Creating a subset Using Indexing operator

```python
subset2=data[["Post Month","Post Hour","Paid"]]
subset2
```

## i) Shape of subset

```python
subset2.shape
```

## j) Subset containing all rows

```python
subset3=data.iloc[:,7:14]
subset3
```

## k) Shape of subset

```python
subset3.shape
```

## l) Creating subset using loc

```python
subset4=data.loc[0:199]
subset4
```

## m) Shape of subset


```python
subset4.shape
```

## n) Subset containing half no. of rows

```python
subset5=data.loc[200:]
subset5
```

## p) Shape of subset

```python
subset5.shape
```

## q) Merge subset using merge function

```python
merge_subset_c=subset1.merge(subset3)
merge_subset_c
```

## r) Shape of merge subset

```python
merge_subset_c.shape
```

## s) Merge subset using concat function

```python
merge_subset_r=pd.concat([subset4,subset5])
merge_subset_r
```

## t) Transpose of dataset

```python
transposeData=data.transpose()
transposeData
```

## u) shape of transpose

```python
transposeData.shape
```

## v) Sorting the dataset

```python
sortedData=data.sort_values(by="Post Month")
sortedData
```

## w) reshape the dataset

```python
reshapeData=data.melt(id_vars=["Type","Category"],value_vars=["Post Month","Post Hour"])
reshapeData
```

## x) shape of the reshaped dataset

```python
reshapeData.shape
```
